```python
import pandas as pd
import requests
from tqdm.notebook import tqdm # если прочитать ошибку она легко исправляться и без сложностей

import collections
import random

# тут запрос тот же что и в курсе, я намеренно ничего не менял, ибо всё есть

vacancies = []
vacancy_query = "product analyst"

for page_number in tqdm(range(50)):
    try:
        response = requests.get("https://api.hh.ru/vacancies",
                                params={
                                    "text": vacancy_query,
                                    "page": page_number
}
                               )
        json_response = response.json()

        vacancies += json_response["items"]
    except Exception as ex:
        print(ex)
```

```python
vacancies_dataset = {
    "vacancy_name": [],
    "employer_name": [],
    "city": [],
    "salary": []  # тут только дополнение одного параметра
}

for vacancy in vacancies:
    vacancies_dataset["vacancy_name"].append(vacancy["name"])
    vacancies_dataset["employer_name"].append(vacancy["employer"]["name"])
    
    try:
        #vacancies_dataset["city"].append(vacancy.get("address", {}).get("city", "no city"))
        vacancies_dataset["city"].append(vacancy["area"]["name"])
    except:
        vacancies_dataset["city"].append("no city")
    try:
        currency = vacancy["salary"]["currency"]
        if currency != 'RUR': # исключение всего что не рубль
            vacancies_dataset["salary"].append(None)
            continue
    except:
        vacancies_dataset["salary"].append(None)
        continue
    try:
        sfrom = vacancy["salary"]["from"]
        if sfrom is None:
            sfrom = 0
    except:
        sfrom = 0
    try:
        sto = vacancy["salary"]["to"]
        if sto is None:
            sto = 0
    except:
        sto = 0
    if sfrom == 0: # просто прибираем значения
        vacancies_dataset["salary"].append(sto)
    elif sto == 0:
        vacancies_dataset["salary"].append(sfrom)
    elif sfrom + sto == 0:
        vacancies_dataset["salary"].append(None)
    else:
        vacancies_dataset["salary"].append((sto+sfrom)/2)
    
vacancies_dataset = pd.DataFrame(vacancies_dataset)
```
